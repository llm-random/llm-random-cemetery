```python
import numpy as np
import matplotlib.pyplot as plt

def find_parabola_coefficients(x1, y1, x2, y2, x3, y3):
    # Setup the system of equations
    A = np.array([
        [x1**2, x1, 1],
        [x2**2, x2, 1],
        [x3**2, x3, 1]
    ])
    B = np.array([y1, y2, y3])
    
    # Solve for coefficients a, b, c
    a, b, c = np.linalg.solve(A, B)
    return a, b, c

def plot_parabola(a, b, c, x1, x2, x3):
    # Generate a range of x values for plotting
    x_values = np.linspace(min(x1, x2, x3) - 1, max(x1, x2, x3) + 1, 500)
    # Calculate corresponding y values
    y_values = a * x_values**2 + b * x_values + c
    
    # Plot the parabola
    plt.plot(x_values, y_values, label='y = {:.2f}x^2 + {:.2f}x + {:.2f}'.format(a, b, c))
    # Plot the original points
    plt.scatter([x1, x2, x3], [y1, y2, y3], color='red', label='Given Points')
    
    # Annotate the points
    plt.text(x1, y1, f'({x1}, {y1})', fontsize=12, ha='right')
    plt.text(x2, y2, f'({x2}, {y2})', fontsize=12, ha='right')
    plt.text(x3, y3, f'({x3}, {y3})', fontsize=12, ha='right')
    
    # Labels and title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Parabola through Three Points')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example points
x1, y1 = np.log(1), 2.806
x2, y2 = np.log(2), 2.774
x3, y3 = np.log(5), 2.807

x1, y1 = np.log(1), 2.809
x2, y2 = np.log(2), 2.732
x3, y3 = np.log(5), 2.776

# Find the coefficients of the parabola
a, b, c = find_parabola_coefficients(x1, y1, x2, y2, x3, y3)

# Plot the parabola
plot_parabola(a, b, c, x1, x2, x3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def find_parabola_coefficients(x1, y1, x2, y2, x3, y3):
    # Setup the system of equations
    A = np.array([
        [x1**2, x1, 1],
        [x2**2, x2, 1],
        [x3**2, x3, 1]
    ])
    B = np.array([y1, y2, y3])
    
    # Solve for coefficients a, b, c
    a, b, c = np.linalg.solve(A, B)
    return a, b, c

def plot_parabolas(params_list):
    plt.figure(figsize=(10, 6))
    
    for params in params_list:
        a, b, c, x1, x2, x3 = params
        # Generate a range of x values for plotting
        x_values = np.linspace(min(x1, x2, x3) - 1, max(x1, x2, x3) + 1, 500)
        # Calculate corresponding y values
        y_values = a * x_values**2 + b * x_values + c

        # Plot the parabola
        plt.plot(x_values, y_values, label='y = {:.2f}x^2 + {:.2f}x + {:.2f}'.format(a, b, c))
        # Plot the original points
        plt.scatter([x1, x2, x3], [y1, y2, y3], color='red')

    # Labels and title
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Two Parabolas')
    plt.legend()
    plt.grid(True)
    plt.show()

# First set of points
x1, y1 = np.log(1), 2.806
x2, y2 = np.log(2), 2.774
x3, y3 = np.log(5), 2.807
a1, b1, c1 = find_parabola_coefficients(x1, y1, x2, y2, x3, y3)

# Second set of points
x1, y1 = np.log(1), 2.809
x2, y2 = np.log(2), 2.732
x3, y3 = np.log(5), 2.776
a2, b2, c2 = find_parabola_coefficients(x1, y1, x2, y2, x3, y3)

# Plot both parabolas
plot_parabolas([(a1, b1, c1, np.log(1), np.log(2), np.log(5)), 
                (a2, b2, c2, np.log(1), np.log(2), np.log(5))])
```
